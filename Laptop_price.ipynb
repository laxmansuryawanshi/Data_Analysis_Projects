```python
for i in cc:
    ssk=df_laptop[i].skew().round()
    print(ssk)
    if ssk==0 or ssk==-0:
        ("There is Normal Standard Deviation ")
        m=df_laptop[i].mean()
        print(m)
        s=df_laptop[i].std()
        ll=m-3*s
        ul=m+3*s
        df_laptop[i].map(lambda x: ll if x<ll else ul if x>ul else x)
    else:
        ("There is the Skewed Distribution")
        print("Inter Qunatile Range")
        q1=df_laptop[i].quantile(0.25)
        print(q1)
        q3=df_laptop[i].quantile(0.75)
        print(q3)
        iqr=q3-q1
        lll=q1-(1.5*iqr)
        ull=q3+(1.5*iqr)
        df_laptop[i].map(lambda x: lll if x<lll else ull if x>ull else x)

```
