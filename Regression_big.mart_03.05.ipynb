## Error removing with the help of IQR( Inter Quantile Range) with TIMMING

```python
q1=dfmart['Item_Visibility'].quantile(0.25)
q3=dfmart['Item_Visibility'].quantile(0.75)
iqr=q3-q1
ll=q1-1.5*iqr
ul=q3+1.5*iqr
print(f' q1 ={q1},q3 ={q3}, iqr ={iqr}, ll= {ll}, ul= {ul}')
```

```python
(dfmart['Item_Visibility']<ll) | (dfmart['Item_Visibility']>ul)
```

```python
dfmart[(dfmart['Item_Visibility']<ll) | (dfmart['Item_Visibility']>ul)]
```

```python
dfmart=dfmart[(dfmart['Item_Visibility']>=ll) &  (dfmart['Item_Visibility']<=ul)]
dfmart
```

```python
sns.boxplot(dfmart)
plt.show()
```

```python
q11=dfmart['Item_Outlet_Sales'].quantile(0.25)
q33=dfmart['Item_Outlet_Sales'].quantile(0.75)
iqr1=q33-q11
lll=q11-1.5*iqr1
ull=q33+1.5*iqr1
print(f' q11 ={q11},q33 ={q33}, iqr ={iqr}, lll= {lll}, ull= {ull}')
```

```python
(dfmart['Item_Outlet_Sales']<lll) | (dfmart['Item_Outlet_Sales']>ull)
```
